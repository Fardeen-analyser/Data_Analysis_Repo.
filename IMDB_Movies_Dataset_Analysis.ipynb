In Which Year There Was The Highest Average Voting

```python
fig, ax = plt.subplots(ncols=1,figsize=(8,6))
sns.barplot(x='Year',y='Votes',data=data , palette = "dark")
plt.title("votes by year",fontsize=15)
plt.xlabel("Years",fontsize=13)
plt.ylabel("Votes",fontsize=13)
for bars in ax.containers:
   ax.bar_label(bars)
plt.show()
```

In Which Year There Was The Highest Average Revenue

```python
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen'})
fig, ax = plt.subplots(ncols=1,figsize=(8,6))
sns.barplot(x='Year',y='Revenue (Millions)',data=data , palette= "dark", legend=True)
plt.title("Highest Average Revenue",fontsize=15)
plt.xlabel("Years",fontsize=13)
plt.ylabel("Revenue (Millions)",fontsize=13)
for bars in ax.containers:
   ax.bar_label(bars)
plt.show()
```

Number of Movies Per Year

```python
data['Year'].value_counts()
sns.countplot(x='Year',data=data, palette = "dark")
plt.title("Number of movies per year",fontsize=15)
plt.xlabel("Year",fontsize=13)
plt.ylabel("Count", fontsize=13)
plt.show()
```

Display Top 10 Highest Rated Movie Titles And its Directors

```python
highest_rated = data.nlargest(10, 'Rating')[['Title','Director','Rating']].set_index('Title')
highest_rated
```

```python
fig, ax = plt.subplots(ncols=1,figsize=(8,6))
sns.barplot(data = highest_rated, y = highest_rated.index, x = 'Rating', hue = 'Director', dodge = False)
plt.legend(bbox_to_anchor=(1.05,1), loc = 2)
for bars in ax.containers:
   ax.bar_label(bars)
plt.show()
```
